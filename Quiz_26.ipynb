```python
import pandas as pa
import seaborn as sns

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')
```

```python
sns.boxplot(data = df.calories)
```

```python
ax = sns.boxplot(data = df, x='workout_type', y='calories')
ax.set_xticklabels(ax.get_xticklabels(),rotation = 45)
```

```python
ax = sns.violinplot(x=df['calories'])
```
