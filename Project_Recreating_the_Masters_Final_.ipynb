##(e) The data is referenced and loaded properly.

Let's load our data set first.

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/almsalahuddin/Data-Processing-and-Visualization/main/3467260/musicdata.csv')
df.head()
```

Next, I will fill all the 'NaN' value with the mean of the 'value_actual' column using fillna function.

```python
df.value_actual.fillna(value=df.value_actual.mean(), inplace=True)
df.head()
```

There are old and adjusted value of the each format and year. So, I will choose to work using only adjusted value.

```python
new=df[df['metric']=='Value (Adjusted)']
new
```

All the references I included at end of this page.

```python
import plotly.express as ax
fig = ax.line(new,  x='year', y='value_actual', color='format')

fig.update_traces(textposition='top center')

fig.update_layout(
    height=800,
    title_text='US music sales by format (inflation-adjusted)'
)
fig.show()
```

Next, I will set a title of the graph and annotate all the lines as a format.

```python
import seaborn as sns
ax=sns.lineplot(data=new, x='year', y='value_actual', hue='format', legend=False)
ax.set(title = "US music sales by format (inflation-adjusted)")
ax.annotate('COMPACT DISK',[1997,20000])
ax.annotate('VINYL',[1977,10000])
ax.annotate('8-TRACK',[1977,4000])
ax.annotate('CASSETTE',[1986,7800])
ax.annotate('OTHER',[2006,1500])
ax.annotate('DOWNLOAD',[2010,2000])
ax.annotate('STREAMING',[2017,6500])
```

Above graph is good but it looks messy because its too small. Let's make this graph liitle bigger using plt.figure function so that it looks cleaner.

```python
import seaborn as sns
plt.figure(figsize=(15,8))
ax=sns.lineplot(data=new, x='year', y='value_actual', hue='format', legend=False)
ax.set(title = "US music sales by format (inflation-adjusted)")
ax.annotate('COMPACT DISK',[1997,20000])
ax.annotate('VINYL',[1977,10000])
ax.annotate('8-TRACK',[1977,4000])
ax.annotate('CASSETTE',[1986,7800])
ax.annotate('OTHER',[2006,1500])
ax.annotate('DOWNLOAD',[2010,2000])
ax.annotate('STREAMING',[2017,6500])
```
