```python
import pandas as pa
df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv')
```

#Barplot

```python
df.groupby('neighbourhood_group').price.agg('max').plot(kind = 'bar',
                                                 title = 'Max price by neighbourhood group', 
                                                 ylabel= 'Price')
```

#Histogram

```python
df.groupby('neighbourhood_group').price.plot.hist(alpha = .7,bins= 500, xlim=[0,1000], title = 'Price of Air B&B broken down by neighbourhood_group', legend = True)
```

```python
import seaborn as sns
df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')
```

#Scatterplot

```python
ax = sns.scatterplot(data = df, 
                x = 'distance', 
                y = 'time',
                hue = 'workout_type'
                )

sns.move_legend(ax, "upper right", bbox_to_anchor=(-.2, 1))
```
