```python
import pandas as pa
df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv')
df.head()
```

```python
df.isnull().sum(axis = 0)
```

```python
df.isnull().sum().sum() 
```

```python
df.reviews_per_month.fillna(value=df.reviews_per_month.mean(), inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.last_review .fillna(value = 10/2/2022, inplace =True)
df
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum() 
```

```python
import numpy as np
df.name = df.name.fillna(df.groupby('host_name').name.transform(lambda x: next(iter(x.mode()), np.nan)))
df
```

```python
df.host_name.fillna(value = 1, inplace =True)
df
```

```python
df.name.fillna(value = 1, inplace =True)
df
```
