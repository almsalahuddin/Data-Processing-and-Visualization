```python
import pandas as pa
df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv')
df.head()
```

Lets find out all the null entries in the dataset.

```python
df.isnull().sum(axis = 0)
```

```python
df.isnull().sum().sum() 
```

As we can see above, there is total of 20141 null entries in the entire dataset.

Now, I will replace last review date to today's date.

```python
df.last_review .fillna(value = 3/4/2022, inplace =True)
df
```

Next, I will check if there is still any null values or not!

```python
df.isnull().sum()
```

Above, we can see that null values for last_review went to 0 that means I was able to replace it successfully. 

Now, I will work on replacing the rest of the null datas.

```python
df.reviews_per_month.fillna(value=df.reviews_per_month.mean(), inplace=True)
df.head()
```

```python
df.isnull().sum()
```

Above, we only have 37 null entries left to replace. Let's replace those to our next part.

```python
import numpy as np
df.name = df.name.fillna(df.groupby('host_name').name.transform(lambda x: next(iter(x.mode()), np.nan)))
df
```

```python
df.isnull().sum()
```

Above we can see still we have 2 null entries for "name" dataset. Lst function didn't work properly for some reason. I will try different method to replace rest.

```python
df.host_name.fillna(value = 'no_name', inplace =True)
df
```

Finally, I will also replace name with no_name.

```python
df.name.fillna(value = 'no_name', inplace =True)
df
```

Now, I will check if we still have any null enteries or everything successfully replaced.

```python
df.isnull().sum()
```

Yoooo!!! No more null entries left. Mission accomplished!
